```python
df_edgesog = pd.read_csv('edge-listsog.csv')
df_math_edges = df_edgesog[df_edgesog['TaxonomyValue'] == 'Mathematics'].copy()
df_math_stats_or_edges = df_edgesog[(df_edgesog['TaxonomyValue'] == 'Mathematics') |(df_edgesog['TaxonomyValue'] == 'Statistics') | (df_edgesog['TaxonomyValue'] == 'Operations Reseach')].copy()
df_stats_edges = df_edgesog[df_edgesog['TaxonomyValue'] == 'Statistics'].copy()
df_or_edges = df_edgesog[df_edgesog['TaxonomyValue'] == 'Operations Research'].copy()
```

```python
#MATH
df_math_edges['Percentage of Women'] = df_math_edges['Women']/(df_math_edges['Men'] + df_math_edges['Women'])*100
#when we have divide by 0, NaN issues, replace with 0 - CAN CHANGE IF NECESSARY
df_math_edges.replace(np.nan, 0.0)
df_math_edges
```

```python
#MATH/OR/STATS
df_math_stats_or_edges['Percentage of Women'] = df_math_stats_or_edges['Women']/(df_math_stats_or_edges['Men'] + df_math_stats_or_edges['Women'])*100
#when we have divide by 0, NaN issues, replace with 0 - CAN CHANGE IF NECESSARY
df_math_stats_or_edges.replace(np.nan, 0.0)
df_math_stats_or_edges

df_stats_edges['Percentage of Women'] = df_stats_edges['Women']/(df_stats_edges['Men'] + df_stats_edges['Women'])*100
#when we have divide by 0, NaN issues, replace with 0 - CAN CHANGE IF NECESSARY
df_stats_edges.replace(np.nan, 0.0)

df_or_edges['Percentage of Women'] = df_or_edges['Women']/(df_or_edges['Men'] + df_or_edges['Women'])*100
#when we have divide by 0, NaN issues, replace with 0 - CAN CHANGE IF NECESSARY
df_or_edges.replace(np.nan, 0.0)
df_or_edges
```

```python
def OverallInstitutionCount(df):
    institution = df['InstitutionName'].unique()
    total_men = np.empty(institution.shape)
    total_women = np.empty(institution.shape)
    total_percentage_of_women = np.empty(institution.shape)

    for i in range(len(institution)):
        df_temp = df[df['InstitutionName'] == institution[i]]
        total_women[i] = np.sum(df_temp['Women'])
        total_men[i] = np.sum(df_temp['Men'])
        total_percentage_of_women[i] = total_women[i] / (total_women[i] + total_men[i]) * 100
    
    d = {'InstitutionName': institution, 'Total Men': total_men, 'Total Women': total_women, 'Total Percentage of Women': total_percentage_of_women}
    df_data_by_institution = pd.DataFrame(data=d)
    return df_data_by_institution
```

```python
df_data_by_institution_math = OverallInstitutionCount(df_math_edges)
df_data_by_institution_msor = OverallInstitutionCount(df_math_stats_or_edges)
df_data_by_institution_stats = OverallInstitutionCount(df_stats_edges)
df_data_by_institution_or = OverallInstitutionCount(df_or_edges)

```
